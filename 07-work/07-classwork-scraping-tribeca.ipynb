```python
import requests
from bs4 import BeautifulSoup
url_letter = 'a'
url_number = '1'
my_url = "https://www.tribecafilm.com/filmguide?letter=" + url_letter + "&page=" + url_number
raw_html = requests.get(my_url).content
soup_doc = BeautifulSoup(raw_html, "html.parser")
print(type(soup_doc))
```

```python
### TEST VIEWS OF DATA
# raw_html
# print(soup_doc)
print(soup_doc.prettify())
```

```python
###  Site code is very clean, each film or page row has been called an "article" in the code
 
first_row = soup_doc.find_all('article')[1]
```

```python
all_rows = soup_doc.find_all('article')
# all_rows
```

```python
###  We'll convert the various fields of the article text into a dictionary for each film, using the first as our template
this_film = {}
film_info = first_row.find(class_='left')
this_film['title'] = film_info.h1.string
headers = film_info.find(class_="headerTags").find_all('a')
head_list = []
for head in headers:
    head_list.append(head.string)
this_film
```

```python
###  Then that technique of making a dictionary will feed every film into a list of dictionaries
all_films = []
for my_row in all_rows:
    this_film = {}
    film_info = my_row.find(class_='left')
    this_film['title'] = film_info.h1.string
    headers = film_info.find(class_="headerTags").find_all('a')
    head_list = []
    for head in headers:
        head_list.append(head.string)
    this_film['headers'] = head_list
    all_films.append(this_film)
```
